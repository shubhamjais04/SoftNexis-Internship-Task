# Step 1: Data Ingestion
Loading the CSV file and taking a first look at the data.

```python
# importing libraries
import numpy as np
import pandas as pd
print("libraries imported successfully!")
```

```python
df = pd.read_csv('customers-1000.csv')
print("Dataset loaded sucessfully!")
```

```python
# shape of the dataset
print("shape:", df.shape)
df.head()
```

## Initial Data Inspection
Checking column names, data types, and basic statistics before cleaning.


```python
df.info()
```

```python
df.describe(include='all')
```

# Step 2: Deduplication
Duplicate rows can affect analysis results, so I'll check and remove them.

```python
print("Duplicate rows found:", df.duplicated().sum())
```

```python
# drop duplicates
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

# Step 3: Column Management
The `Index` column is just a serial number, pandas handles that automatically.  
`Phone 2` is a secondary number which is not needed for analysis. Dropping both.

```python
df = df.drop(columns=['Index', 'Phone 2'])
print("Remaining columns:", df.columns.tolist())
```

## Renaming Columns
Renaming columns to lowercase with underscores — cleaner and easier to work with in code.

```python
df.columns = ['customer_id', 'first_name', 'last_name', 'company', 
              'city', 'country', 'phone', 'email', 'subscription_date', 'website']
print("Renamed columns:", df.columns.tolist())
```

# Step 4: Missing Value Handling
Let's see if any column has missing/null values.

```python
print("Missing values per column:")
print(df.isna().sum())
```

```python
print("Total missing values:", df.isna().sum().sum())
```

# Step 5: Data Type Correction 
The `subscription_date` column is stored as a string. Converting it to datetime format

```python
print("Before:", df['subscription_date'].dtype)
```

```python
df['subscription_date'] = pd.to_datetime(df['subscription_date'], errors='coerce')
print("After:", df['subscription_date'].dtype)
```

# Step 6: Format Standardization 
I noticed the following issues:  
- `customer_id` has mixed upper/lowercase letters  
- `email` is not consistently lowercase  
- `first_name`, `last_name`, `city`, `country` need proper title casing  
- Extra whitespace needs to be removed from all text columns

```python
df['customer_id'] = df['customer_id'].str.upper().str.strip()
df['email'] = df['email'].str.lower().str.strip()
df['first_name'] = df['first_name'].str.title().str.strip()
df['last_name'] = df['last_name'].str.title().str.strip()
df['city'] = df['city'].str.title().str.strip()
df['country'] = df['country'].str.title().str.strip()
df['company'] = df['company'].str.strip()
df['website'] = df['website'].str.strip()

print("Text columns standardized!")
df[['customer_id', 'email', 'first_name', 'last_name', 'city', 'country', 'subscription_date', 'company', 'website']].head()
```

## Extracting Subscription Year
Since the date is now properly formatted, I'll extract the year into a  
new column — useful for yearly analysis.

```python
df['subscription_year'] = df['subscription_date'].dt.year
print("Subscription year range:", df['subscription_year'].min(), "to", df['subscription_year'].max())
df[['subscription_date', 'subscription_year']].head()
```

## Final Result
Running final checks to confirm the data is properly cleaned.

```python
print("Final shape:", df.shape)

print("\nMissing values after cleaning:")
print(df.isna().sum())

print("\nDuplicates after cleaning:", df.duplicated().sum())

print("\nData types:")
print(df.dtypes)
```

# Step 7: Saving the Cleaned Dataset
Saving the final cleaned data as a new CSV file.

```python
df.to_csv('cleaned_customers.csv', index=False)
print("Cleaned dataset saved as 'cleaned_customers.csv'!")
df.head()
```
